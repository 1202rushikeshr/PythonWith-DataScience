```python
import numpy as np
import pandas as pd
iris_df=pd.read_csv('C:/Users/1202r/Iris.csv')
iris_df.shape

X=iris_df.iloc[:,1:5].values
y=iris_df['Species'].values
```

```python
y
```

## Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
iris_df['Species_encoded']=encoder.fit_transform(iris_df['Species'])
iris_df.tail(51)
```

```python
iris_df['Species_encoded'].value_counts()
```

## LOGESTIC REGRESSION

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.3,random_state=10)
from sklearn.linear_model import LogisticRegression
logRegModel=LogisticRegression()
logRegModel.fit(X_train,y_train)
y_preds=logRegModel.predict(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_preds))
```

```python
from sklearn.ensemble import RandomForestClassifier
rfClassifierModel=RandomForestClassifier(n_estimators=5)
rfClassifierModel.fit(X_train,y_train)
y_preds_rf=rfClassifierModel.predict(X_test)
print(classification_report(y_test,y_preds_rf))
```
