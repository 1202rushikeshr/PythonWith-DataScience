```python
#15. Tenese School Data Analysis
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Phase-1 : COLLECT DATA
tn_schools_df=pd.read_csv('middle_tn_schools.csv')
tn_schools_df.describe()
tn_schools_df.info()
tn_schools_df.isnull().sum()

#Phase 2: Group Data by School Ratings & Isolates data on ‘reduced_lunch’
grouped_data=tn_schools_df.groupby('school_rating')['reduced_lunch'].describe()
print(grouped_data)
#Phase3:Establisj=hing co-reletion
rl_sr_correlation = tn_schools_df[['reduced_lunch', 'school_rating']].corr()
str_sr_correletion=tn_schools_df[['stu_teach_ratio','school_rating']].corr()
print("ReducedLunch & SchoolRating Correletion:",rl_sr_correlation)
#print("StuTechRatio & SchoolRating Correletion:",str_sr_correletion)

# Phse-4:Scatter plot Reduced_Lunch Vs School_Rating
plt.scatter(tn_schools_df['reduced_lunch'], tn_schools_df['school_rating'])
z = np.polyfit(tn_schools_df['reduced_lunch'], tn_schools_df['school_rating'], 1)
p = np.poly1d(z)
plt.plot(tn_schools_df['reduced_lunch'], p(tn_schools_df['reduced_lunch']), color='red', label='Trend Line')

plt.title("Scatter Plot of School Rating vs Reduced Lunch", fontsize=14)
plt.xlabel("Reduced Lunch (%)", fontsize=12)
plt.ylabel("School Rating", fontsize=12)
plt.legend()
plt.grid(True)
plt.show()
```
