```python
import numpy as np
import pandas as pd
iris_df=pd.read_csv('Iris.csv')
iris_df.head
```

```python
X=iris_df.iloc[:,1:5].values
y=iris_df['Species'].values
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=10)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtClassifierModel=DecisionTreeClassifier()
dtClassifierModel.fit(X_train,y_train)
y_preds=dtClassifierModel.predict(X_test)
from sklearn.metrics import classification_report
print(classification_report(y_test,y_preds))
```

## To use PCA for dimensionality reduction , Vaules need to be Normalised first

```python
## Normlizing the input values
from sklearn.preprocessing import MinMaxScaler
mmScalerModel=MinMaxScaler()
X_train=mmScalerModel.fit_transform(X_train)
X_test=mmScalerModel.fit_transform(X_test)
X_train
```

## PCA Technique for dimensionality reduction

```python
from sklearn.decomposition import PCA
pcaModel=PCA()
X_train=pcaModel.fit_transform(X_train)
X_test=pcaModel.transform(X_test)
pcaModel.explained_variance_ratio_
```

```python
## So from abhove Array(igen Vector) it seems out of 4 , first 2 values array[0] & array[1] together contributes for 0.95(95%) of total verience 
pcaModel=PCA(n_components=2)  ## considering only 2 values
X_train=pcaModel.fit_transform(X_train)
X_test=pcaModel.transform(X_test)
pcaModel.explained_variance_ratio_
```

```python
dtClassifierModel.fit(X_train,y_train)
y_preds_pca=dtClassifierModel.predict(X_test)
print(classification_report(y_test,y_preds_pca))
```

## Linear Discriminent Analysis(LDA)

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
ldaModel=LDA(n_components=2)
X_train=ldaModel.fit_transform(X_train,y_train)
X_test=ldaModel.transform(X_test)
dtClassifierModel.fit(X_train,y_train)
y_preds_lda=ldaModel.predict(X_test)
print(classification_report(y_test,y_preds_lda))
```
